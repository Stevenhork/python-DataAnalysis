```python
#开始数据清洗
#1、获取有效字段

effectiveField = ["movie_name","director","screenwriter","starring_people","movie_type",
                   "production_country","release_date","movie_duration","movie_rating","movie_num_people",
                  "movie_short_comment","movie_content","movie_all_comment"]
movieDatas_effectiveField = movieDatas_all[effectiveField]
movieDatas_effectiveField
```

```python
movieDatas_effectiveField.columns
```

```python
# 2、字段重命名，避免政治错误

movieDatas_effectiveField.rename(columns={"production_country":"production_region"},inplace=True)
movieDatas_effectiveField.columns
```

```python
#3、清除导演中的非法数据

movieDatas_clean = movieDatas_effectiveField[movieDatas_effectiveField["director"]\
    .apply(lambda x: "纪录片" not in str(x) and "空" not in str(x) and "???" not in str(x))]
movieDatas_clean["director"]
```

```python
# 4、清除编剧的非法数据

#获取编剧非法数据集
movie_types = list(movieDatas_clean["movie_type"].unique())
illegalDatas_screenwriter = movie_types

#将" "换成"/"
illegalDatas_screenwriter = [i.replace(" "," / ")[0:-3] for i in illegalDatas_screenwriter]
illegalDatas_screenwriter

movieDatas_clean = movieDatas_clean[movieDatas_clean["screenwriter"].apply(lambda x: str(x) not in illegalDatas_screenwriter )]
movieDatas_clean = movieDatas_clean[movieDatas_clean["screenwriter"].apply(lambda x:"http" not in str(x) and "地区" not in str(x) and "空" not in str(x) and
                                      "www" not in str(x) and "???" not in str(x) and "未知" not in str(x))]
movieDatas_clean["screenwriter"]
```

```python
#5、清除演员非法数据

movieDatas_clean = movieDatas_clean[movieDatas_clean["starring_people"]\
                 .apply(lambda x: "无数据" not in str(x) and "空" not in str(x) and "???" not in str(x))]
movieDatas_clean["starring_people"]
```

```python
# 6、清除电影类型非法数据
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_type"].apply(lambda x: "无数据" not in x)]
movieDatas_clean["movie_type"]
```

```python
# 7、清除拍摄地区非法数据
movieDatas_clean = movieDatas_clean[movieDatas_clean["production_region"]\
.apply(lambda x: "无数据" not in x and '/' not in x and "语" not in x and "." not in x and "n" not in x and "非中国大陆" not in x)]
movieDatas_clean["production_region"]
```

```python
#将中国与中国大陆的数据进行合并
movieDatas_clean["production_region"].replace("中国","中国大陆",inplace=True)
movieDatas_clean[movieDatas_clean["movie_name"] == "中国"]
```

```python
# 8、清除上映时间的非法数据，并截取年份存入release_year
movieDatas_clean = movieDatas_clean[movieDatas_clean["release_date"].apply(lambda x: "无数据" not in x)]
movieDatas_clean["release_year"] = movieDatas_clean["release_date"].apply(lambda x:x[0:4])
#取年份第一位，对比 ASCII 去除非法字符
movieDatas_clean = movieDatas_clean[movieDatas_clean["release_year"].apply(lambda x:ord(x[0:1]) < 58)]
movieDatas_clean
```

```python
# 9、将获取的年份转为 int
movieDatas_clean["release_year"] = movieDatas_clean["release_year"].astype("int64")
movieDatas_clean["release_year"]
```

```python
# 10、清除影片时长非法数据,并获取影片时长数组，且将其转为 int

#清除非法数据
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_duration"].apply(lambda x: "无数据" not in str(x))]

#获取数字
movieDatas_clean["movie_duration"] = movieDatas_clean["movie_duration"].apply(lambda x: x[0:str(x).find("分")])

# 去除影片时长非法字符，并将其转为 int
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_duration"].apply(lambda x:len(str(x)) > 0)]
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_duration"].apply(lambda x:ord(x[0:1]) < 58)]
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_duration"].apply(lambda x: "-" not in x and "None" not in x and "小时" not in x)]
movieDatas_clean["movie_duration"] = movieDatas_clean["movie_duration"].astype("float")
movieDatas_clean["movie_duration"]
```

```python
# 11、将评分转为 float
movieDatas_clean["movie_rating"].astype("float")
movieDatas_clean["movie_rating"]
```

```python
#12、去除将打分人数非法数据，并将其转为 int
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_num_people"].apply(lambda x: "无数据" not in str(x))]
movieDatas_clean["movie_num_people"].astype("int64")
```

```python
#13、获取短评论数，并将其转为 int
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_short_comment"].apply(lambda x: "无数据" not in str(x))]
movieDatas_clean["movie_short_comment"] = movieDatas_clean["movie_short_comment"].apply(lambda x:x[2:-1])
movieDatas_clean["movie_short_comment"].apply(lambda x : len(str(x)) > 0)
movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_short_comment"].apply(lambda x : len(str(x)) > 0)]
movieDatas_clean["movie_short_comment"] = movieDatas_clean["movie_short_comment"].astype("int64")
movieDatas_clean["movie_short_comment"]
```

```python
# 14、获取评论数，并将其转为 int

movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_content"].apply(lambda x: "无数据" not in str(x))]
movieDatas_clean["movie_content"] = movieDatas_clean["movie_content"].apply(lambda x:x[2:-1])

movieDatas_clean = movieDatas_clean[movieDatas_clean["movie_content"].apply(lambda x : len(str(x)) > 0)]
movieDatas_clean["movie_content"] = movieDatas_clean["movie_content"].astype("int64")
movieDatas_clean["movie_content"]
```

```python
#创建一个指定大小的二维数组
China_movies_date_type_array = [0]*len(China_movieDate)
for i in range(len(China_movies_date_type_array)):
    China_movies_date_type_array[i] = [0]*len(China_movieTypes)

#往数组中填入数据
for i in China_movieDate:
    for j in China_movieTypes:  
        # 遍历对应年内，全部电影的类型
        for k in list(China_movies[China_movies["release_year"] == i]["movie_type"]):
            # k：对应年内，某部电影的类型
            if j in k:
                y = China_movieDate.index(i)
                x = China_movieTypes.index(j)
                China_movies_date_type_array[y][x] += 1
China_movies_date_type_array
```

```python
#获取所有影片的中国导演
director_data = movieDataMerge.copy()
director_all = list(director_data[director_data["production_region"] == "中国大陆"]["director"])
directors = []
for i in director_all:
    arr = i.split(" / ")
    for j in arr:
        if len(j) > 0:
            directors.append(j)
china_directors = list(set(directors))   #所有中国导演的名字
china_directors
```

```python
#获取每个导演的评分总和及电影数目  {"director":[rating,count]}
directorRatingMap = {}
for i in china_directors:
    key = i
    rating = count = 0
    ratingList = list(China_movies[China_movies["director"].apply(lambda x: i in x)]["movie_rating"])
    for j in ratingList:
        if j is not None:
            rating += j
            count += 1
    ratingData = directorRatingMap.get(key)
    if ratingData is not None:
        if ratingData[0] != 0:
            rating += ratingData[0]
        count += ratingData[1]
    directorRatingMap[key] = [rating,count]
directorRatingMap
```

```python
#获取影片综合数据
movieComprehensiveDataList = []

#获取所有电影的名字列表
movieNameList = list(China_movies["movie_name"].unique())

#求每一部电影的 【导演、编剧、演员】 的综合评分及票房
for i in movieNameList:

    #获取该电影的导演列表
    directorList = list(list(China_movies[China_movies["movie_name"] \
                                     == i]["director"])[0].split(" / "))
    directorCapacity_sum = 0   # 该电影的导演能力值总和
    directorCapacity_count = 0  # 该电影导演的个数

    #获取该电影每一位导演的 能力值
    for j in director_capacity.index:
        for k in directorList:
            if j == k:
                #若 k 导演导过 i 影片，则将该导演的能力值加入到导演能力值总数中（directorCapacity_sum）
                directorCapacity = director_capacity["capacity"][k]
                directorCapacity_sum += directorCapacity
                directorCapacity_count += 1
                
    #获取该电影的编剧列表
    screenwriterList = list(list(China_movies[China_movies["movie_name"] \
                                         == i]["screenwriter"])[0].split(" / "))
    screenwriterCapacity_sum = 0   # 该电影的编剧能力值总和
    screenwriterCapacity_count = 0  # 该电影编剧的个数

    #获取该电影每一位编剧的 能力值
    for j in screenwriter_capacity.index:
        for k in screenwriterList:
            if j == k:
                #若 k 编剧编过 i 影片，则将该编剧的能力值加入到编剧能力值总数中（screenwriterCapacity_sum）
                screenwriterCapacity = screenwriter_capacity["capacity"][k]
                screenwriterCapacity_sum += screenwriterCapacity
                screenwriterCapacity_count += 1

    #获取该电影的演员列表
    starringList = list(list(China_movies[China_movies["movie_name"] \
                                          == i]["starring_people"])[0].split(" / "))
    starringCapacity_sum = 0   # 该电影的编剧能力值总和
    starringCapacity_count = 0  # 该电影编剧的个数

    #获取该电影每一位主演的 能力值
    for j in starring_capacity.index:
        for k in starringList:
            if j == k:
                #若 k 主演演过 i 影片，则将该主演的能力值加入到演员能力值总数中（starringCapacity_sum）
                starringCapacity = starring_capacity["capacity"][k]
                starringCapacity_sum += starringCapacity
                starringCapacity_count += 1
    
    #获取该影片的票房
    boxOffice = list(China_movies[China_movies["movie_name"] == i]["movie_box_office(万元)"])[0]
    
    #获取该影片的评分
    rating = list(China_movies[China_movies["movie_name"] == i]["movie_rating"])[0]
    
    #存放一部影片的综合数据
    movieDataList = [directorCapacity_sum,
                     directorCapacity_count,
                     directorCapacity_sum/( 1 if directorCapacity_count == 0 else directorCapacity_count),
                     screenwriterCapacity_sum,
                     screenwriterCapacity_count,
                     screenwriterCapacity_sum/( 1 if screenwriterCapacity_count == 0 else screenwriterCapacity_count),
                     starringCapacity_sum,
                     starringCapacity_count,
                     starringCapacity_sum/( 1 if starringCapacity_count == 0 else starringCapacity_count),
                     rating,
                     boxOffice,
                    ]
    #存放所有影片的综合数据
    movieComprehensiveDataList.append(movieDataList)
```
